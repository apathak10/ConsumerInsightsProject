```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
```

```python
#Note that a csv file has been imported in this notebook
#This csv file has been fetched from Amazon by using an open-source web scraping tool called ParseHub
#As the dataset is not cleaned, it needs to be cleaned for analyzing in Tableau
df_honey_bunches = pd.read_csv('ParseHub_Scraping_Amazon.csv', encoding='utf-8')
```

```python
df_honey_bunches.head()
```

```python
df_honey_bunches.drop_duplicates(inplace=True)
df_honey_bunches.info()
```

```python
df_honey_bunches['User_country']='United States'
```

```python
new_df = df_honey_bunches['Review_date'].str.split('on ', expand=True)
df_honey_bunches['Review_date'] = new_df[1]
```

```python
new_df = df_honey_bunches['Rating'].str.split(' out', expand=True)
df_honey_bunches['Rating'] = new_df[0]
```

```python
df_honey_bunches['Helpful_votes'].fillna(0, inplace=True)
```

```python
new_df = df_honey_bunches['Helpful_votes'].str.split(' people', expand=True)
df_honey_bunches['Helpful_votes'] = new_df[0]
```

```python
df_honey_bunches
```

# Sentiment Analysis

```python
vader = SentimentIntensityAnalyzer()
review_polarity_score = []
for i in df_honey_bunches['Text']:
    review_polarity_score.append(vader.polarity_scores(i)['compound'])
df_honey_bunches['Polarity_score'] = review_polarity_score
df_honey_bunches['Sentiment'] =''
```

```python
mask = (df_honey_bunches['Polarity_score'] > 0.05) 
df_honey_bunches['Sentiment'][mask] = 'Positive'
mask = (df_honey_bunches['Polarity_score'] < -0.05)
df_honey_bunches['Sentiment'][mask] = 'Negative'
mask = (df_honey_bunches['Polarity_score'] >= -0.05) & (df_honey_bunches['Polarity_score'] <= 0.05)
df_honey_bunches['Sentiment'][mask] = 'Neutral'
```
